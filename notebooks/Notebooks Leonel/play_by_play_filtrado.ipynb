```python
import pandas as pd

# Rutas a los archivos
play_by_play_path = "../../Data Cruda/play_by_play.csv"
game_path = "../../Data/Data Cruda/game.csv"

# Leer los datasets
df_play = pd.read_csv(play_by_play_path)
df_game = pd.read_csv(game_path)

# Limpiar nombres de columnas
df_game.columns = df_game.columns.str.strip()

# Convertir columna de fecha
df_game['game_date'] = pd.to_datetime(df_game['game_date'], errors='coerce')

# Merge por game_id incluyendo ambas columnas de equipos
df_merged = df_play.merge(
    df_game[['game_id', 'game_date', 'team_name_home', 'team_name_away']],
    on='game_id',
    how='left'
)

# Filtrar por fecha mayor a 2018-01-01
df_filtrado = df_merged[df_merged['game_date'] > pd.Timestamp('2018-01-01')]

# Lista de equipos a conservar
equipos_filtrados = [
    "Milwaukee Bucks",
    "Philadelphia 76ers",
    "Denver Nuggets",
    "Boston Celtics",
    "Toronto Raptors",
    "Utah Jazz",
    "Golden State Warriors",
    "Miami Heat",
    "LA Clippers",
    "Los Angeles Lakers"
]

# Filtrar si alguno de los dos equipos estÃ¡ en la lista
df_top_equipos = df_filtrado[
    (df_filtrado['team_name_home'].isin(equipos_filtrados)) |
    (df_filtrado['team_name_away'].isin(equipos_filtrados))
]

# Verificar el resultado
print(df_top_equipos[['game_id', 'team_name_home', 'team_name_away', 'game_date']].drop_duplicates().head())

#  Guardar 
df_top_equipos.to_csv("../../Data Limpia/play_by_play_top10_filtrado_home_away.csv", index=False)
```

```python
import pandas as pd

ruta = "../../Data Limpia/play_by_play_top10_filtrado_home_away.csv"
df = pd.read_csv(ruta)
# Detectar outliers con mÃ©todo IQR
def detectar_outliers_iqr(df):
    outliers = {}
    for col in df.select_dtypes(include='number').columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        filtro = (df[col] < (Q1 - 1.5 * IQR)) | (df[col] > (Q3 + 1.5 * IQR))
        cantidad = filtro.sum()
        if cantidad > 0:
            outliers[col] = cantidad
    return outliers

# Aplicar la funciÃ³n y mostrar los resultados en forma de lista
outliers_detectados = detectar_outliers_iqr(df)

print("Outliers detectados por columna:")
for columna, cantidad in outliers_detectados.items():
    print(f"- {columna}: {cantidad}")
```
