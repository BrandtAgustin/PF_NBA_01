```python
# 1. Leer archivos
team = pd.read_csv("../../Data Cruda/team.csv")
team_history = pd.read_csv("../../Data Cruda/team_history.csv")

# 2. Filtrar equipos
equipos_filtrados = [
    "Milwaukee Bucks",
    "Philadelphia 76ers",
    "Denver Nuggets",
    "Boston Celtics",
    "Toronto Raptors",
    "Utah Jazz",
    "Golden State Warriors",
    "Miami Heat",
    "Los Angeles Clippers",
    "Los Angeles Lakers"
]

team_filtrado = team[team['full_name'].isin(equipos_filtrados)].copy()

# 3. Preparar claves
team_filtrado.rename(columns={'id': 'team_id'}, inplace=True)
team_filtrado['team_id'] = team_filtrado['team_id'].astype(str)
team_history['team_id'] = team_history['team_id'].astype(str)

# 4. (Opcional) Agregar los Celtics si no están
if '1610612738' not in team_history['team_id'].values:
    fila_celtics = {
        'team_id': '1610612738',
        'city': 'Boston',
        'nickname': 'Celtics',
        'year_founded': 1946,
        'year_active_till': 2019
    }
    team_history = pd.concat([team_history, pd.DataFrame([fila_celtics])], ignore_index=True)

# 5. Merge solo sobre los equipos filtrados
team_final = team_filtrado.merge(
    team_history[['team_id', 'year_active_till']],
    on='team_id',
    how='left'
)

# 6. Guardar resultado
team_final.to_csv("../../Data Limpia/team_completa.csv", index=False)
```

```python
# Quedarnos solo con la última etapa histórica por team_id
historia_reciente = team_history.sort_values('year_active_till', ascending=False).drop_duplicates('team_id')

# Merge usando solo la versión más actual
team_final = team_filtrado.merge(
    historia_reciente[['team_id', 'year_active_till']],
    on='team_id',
    how='left')
```
