```python
# 📌 Importamos las librerías necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re

# 📌 Ruta del archivo CSV #game_info.csv
ruta = "../../Data/Data Cruda/game_info.csv"

# 📌 Cargamos el CSV
df_inactive = pd.read_csv(ruta)

# 📌 Mostramos estructura general del DataFrame
df_inactive.info()
df_inactive.head()
df_inactive.shape
```

```python
# 📌 Mostrar cantidad de valores nulos por columna
nulos = df_inactive.isnull().sum()
nulos = nulos[nulos > 0].sort_values(ascending=False)

print("📊 Valores nulos por columna:")
print(nulos)
```

```python
# 📌 Mostrar filas que contienen al menos un valor nulo
filas_con_nulos = df_inactive[df_inactive.isnull().any(axis=1)]

# 📌 Ver las primeras filas con nulos
filas_con_nulos.head()
```

```python
# 📌 Eliminar la fila que contiene valores nulos en attendance o game_time
df_inactive = df_inactive[~(df_inactive['attendance'].isnull() | df_inactive['game_time'].isnull())]
```

```python
# 📌 Mostrar filas que contienen al menos un valor nulo
filas_con_nulos = df_inactive[df_inactive.isnull().any(axis=1)]

# 📌 Ver las primeras filas con nulos
filas_con_nulos.head()
```

```python
# 📌 Verificar si hay filas duplicadas en todo el DataFrame
duplicados = df_inactive.duplicated()

# 📌 Mostrar cuántas filas están duplicadas
print(f"Filas duplicadas: {duplicados.sum()}")

# 📌 Mostrar las filas duplicadas
df_inactive[duplicados]
```

```python
# 📌 Paso 1: Detectar filas duplicadas (todas las columnas)
duplicados = df_inactive[df_inactive.duplicated(keep=False)]

# 📌 Paso 2: Contar cuántas veces se repite cada fila duplicada
resumen_duplicados = duplicados.value_counts().reset_index(name='Repeticiones')

# 📌 Paso 3: Mostrar resumen
print(f"🔁 Total de filas duplicadas: {len(duplicados)}")
print("📋 Resumen de duplicados (filas idénticas):")
print(resumen_duplicados.head())

# 📌 Paso 4: Mostrar tabla de duplicados
duplicados
```

```python
# 📌 Paso 1: Detectar duplicados en la columna 'game_id'
duplicados_id = df_inactive[df_inactive.duplicated(subset='game_id', keep=False)]

# 📌 Paso 2: Mostrar los duplicados encontrados
print(f"🔁 Total de filas con id duplicado: {len(duplicados_id)}")
duplicados_id.head()
```

```python
# ✔️ Convertir a datetime
df_inactive['game_date'] = pd.to_datetime(df_inactive['game_date'], errors='coerce')

# ✔️ Filtrar por fecha >= 2018-01-01
df_inactive = df_inactive[df_inactive['game_date'] >= '2018-01-01']
```

```python
# 📌 Exportar el DataFrame limpio a CSV en la carpeta del proyecto
ruta_salida = "../../Data/Data Limpia/game_info.csv"
df_inactive.to_csv(ruta_salida, sep=',', index=False, decimal='.')

print("✅ Archivo exportado correctamente como game_info.csv")
```
